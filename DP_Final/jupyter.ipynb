#Importing required packages

```python
import numpy as nm
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
print("imported")
```

#Reading the CSV file and displaying its info.

```python
sp = pd.read_csv('CoffeeAndCodeLT2018.csv')
print(sp.columns)
print("----------------------------------------------------------------------------")
print("----------------------------------------------------------------------------")
sp.info()
```

```python
sp.describe()
```

1. Type of Coffee mostly preferred?

```python
variable = 'CoffeeType'
var = sp[variable]
    
#visualization
plt.figure(figsize=(20,4))
sns.countplot(x=var, palette="dark", order=var.value_counts().index)

```

2. CoffeeTime vs CoffeeSolveBugs - while differntiating on Gender

```python
plt.figure(figsize=(20,4))
sns.swarmplot(y='CoffeeTime', x = 'CoffeeSolveBugs', data = sp,hue='Gender')
plt.show()
```

```python
def count_plot(variable):
    
    var = sp[variable]
    
    #visualization
    plt.figure(figsize=(10,4))
    sns.countplot(x=var, palette="dark",order=var.value_counts().index)
    plt.xticks(rotation=45)
    plt.ylabel("Frequency")
    plt.title(variable)
    print("{}".format(var.value_counts()))
    plt.show()
categorical = ["CoffeeTime", "CodingWithoutCoffee", "CoffeeSolveBugs", "AgeRange"]
for i in categorical:
    count_plot(i)
```

4. Correlation between CodingHours and CoffeeCupsPerDay

```python
plt.figure(figsize=(10,4))
corr_matrix = sp.corr()
sns.heatmap(corr_matrix, annot = True)
plt.show()

plt.figure(figsize=(20,4)) 
sns.pairplot(sp)
```

5. Coffee cups per day VS CoffeeTime

```python
ccpd=sp["CoffeeCupsPerDay"]
age=sp["CoffeeTime"]
plt.scatter(ccpd,age)
plt.xlabel('CoffeeCupsPerDay')
plt.ylabel('CoffeeTime')
plt.title('CoffeeTime vs CoffeeCupsPerDay')
plt.show()
```
